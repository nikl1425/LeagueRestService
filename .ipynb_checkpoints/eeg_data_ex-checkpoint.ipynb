```python
def get_window(index, channel, data, fq, size=4, overlap=2):
    start = int(index*(size-overlap)*fq)
    end = int(start + size * fq)
    #print(start,end)
    ts_start = data['timestamp'][start:start+1].tolist()[0]
    #print(datetime.fromtimestamp(ts_start/1000))
    return (data[channel][start:end].tolist(), datetime.fromtimestamp(ts_start/1000))
    
#w = get_window(100, 'CH-1', df)
#plt.plot(w)
#plt.show()
#get_window(0, 'CH-1', sz, freq)
```

```python
for x in range(60,70):
    (w,t) = get_window(x, 'CH-1', sz, freq)
    w = np.array(w)
    #y = denoise_wavelet(w, method='VisuShrink', wavelet='db6', mode='soft', wavelet_levels=3, rescale_sigma='True')
    y = denoise_wavelet(w, method='BayesShrink', wavelet='db6', mode='soft', wavelet_levels=3, rescale_sigma='True')
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(4)
    fig.suptitle(t)
    ax1.plot(w)
    s,f,t,im = ax2.specgram(w,Fs=freq,cmap='jet')
    #print(im)
    #s,f,t,im = ax2.specgram(w,Fs=freq,cmap='jet')
    ax3.plot(y)
    s,f,t,im = ax4.specgram(y,Fs=freq,cmap='jet')
```
